##Types of Graphs

```python
import matplotlib.pyplot as plt
import numpy as np

# Basic Line Plot
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])

plt.plot(x,y)
plt.title('Bacis Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()


```

```python
#Bar Graph
x = ['A', 'B', 'C']
y = [5, 7, 3]

plt.bar(x,y)
plt.title('Bar Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Histogram
data = np.random.randn(1000)
plt.hist(data, bins=20, color='orange', edgecolor='black')
plt.title("Histogram")
plt.show()
```

```python
#Pie Chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Customizations
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])

plt.plot(x,y,color = 'green', linestyle = '--', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title ('Customized Line Plot')

```

```python
plt.plot(x,y,label = 'Line 1')
plt.legend()
plt.show()
```

```python
#Changing Line Style Color Marker
plt.plot(x, y, color='green', linestyle='--', marker='o', linewidth=8, markersize=8)
plt.title("Customized Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
#Title & Label Formatting
plt.plot(x, y)
plt.title("Parabola Plot", fontsize=16, color='darkred')
plt.xlabel("Input", fontsize=12, color='blue')
plt.ylabel("Output", fontsize=12, color='blue')
plt.grid(True)
plt.show()
```

```python
#Grid Customization
plt.plot(x, y)
plt.grid(True, color='gray', linestyle=':', linewidth=1)
plt.show()
```

```python
#Legend With Shadow
plt.plot(x, y, label='x squared', color='green')
plt.legend(loc='upper left', fontsize=10, shadow=True)
plt.show()
```

```python
# Axis Limits
plt.plot(x, y)
plt.xlim(0, 5)   # x-axis range
plt.ylim(0, 20)  # y-axis range
plt.show()
```

```python
# SubPlot

fig, axs = plt.subplots(2, 2) #Means 2 Figs in 2 rows as shown below

axs[0, 0].plot(x, y) #Line Bar
axs[0, 1].bar(['A', 'B'], [5, 7]) #Bar chart
axs[1, 0].hist(data) #Histogram
axs[1, 1].pie([10, 20, 30], labels=['X', 'Y', 'Z']) #Pie Chart
fig.patch.set_facecolor('lightblue') #Background Color
plt.tight_layout() # Formats the space
plt.savefig("plot1.png")
plt.show()
```
