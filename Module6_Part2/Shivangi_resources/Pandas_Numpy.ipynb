## Day 5 - Tuesday
Outliner Detection & Treatment

Outlier Detection: Process of identifying diff cat of data from majority of data.

Methods:
Standard deviation
IQR (InterQuartile Range)
Z-Score
Clustering
Isolation Forest

Choice of detection technique depends on data statistics.


```python
import pandas as pd
import numpy as np
from scipy import stats

#Original Salary dataset
Salary = np.array([40000, 42000, 41500, 39800, 400000])

#The z-score is used to detect outliers like 400000
#We need to find the max of non-outliers salaries, here assumed to be 42000
#Replace the outlier with the max value
#Calculate z-score

z_scores = stats.zscore(Salary)
print(z_scores)

#Define a lower threshold to catch outliers in small datasets.
threshold = 1.0

#Identify outlier indices
outlier_indices = np.where(np.abs(z_scores)>threshold)[0]
print("outlier_indices ",outlier_indices)

#Get max outlier salary
non_outliers = Salary[np.abs(z_scores)<=threshold]
max_non_outlier = non_outliers.max()

#Replace outliers with Max Non-Outlier value
salary_cleaned = Salary.copy()
salary_cleaned[outlier_indices] = max_non_outlier
print("Cleaned Salaries: ", salary_cleaned)
```

The Z-score method identifies outliers based on the number of standard deviations a data point is from the mean. A Z-score is calculated as:

z =  (DataPoints - Mean)  Standard Deviation
where:

X is the data point,
 is the mean, and
 is the standard deviation.

Pros and Cons

Pros: Easy to implement, works well for normally distributed data.
Cons: Not ideal for skewed data, sensitive to extreme outliers.

```python
import numpy as np
import pandas as pd

# Sample data
data = {'values': [12, 15, 14, 10, 102, 13, 11, 14, 13, 15, 110]}
df = pd.DataFrame(data)
# Calculate Z-scores
df['z_score'] = (df['values'] - df['values'].mean()) / df['values'].std()
# Identify outliers
outliers = df[np.abs(df['z_score']) > 1.6]
print("Outliers:\n", outliers)
```

```python
#Replace outliers with Max Non-Outlier value
median_value = df['values'].median()
df.loc[np.abs(df['z_score']) > threshold, 'values'] = median_value

# Drop z_score column after treatment
df.drop('z_score', axis=1, inplace=True)

print("\nAfter Treatment:\n", df)
```

```python
#Normalization: Is a process of scaling numeric values to a common range ussually between 0 & 1.
#Min-Max Normalization

'''X(norm) = X-x(min)/(x(max)-x(min))

where:
X min is the minimum value of feature
X max is the maximum value of feature '''

#Normalize Data below (Min-Max Normalization):

import numpy as np
# Original exam scores
scores = np.array([60, 75, 88, 92, 55, 70])

# Min-Max normalization
normalized_scores = (scores - scores.min()) / (scores.max() - scores.min())
'''min-max normalization:
scores.min()  55
scores.max()  92
So the range is 92 - 55 = 37'''

#For 60  (6055)/37=0.1351
#For 75  (7555)/37=0.5405. Use this for rest of scores.

print("Original Scores:", scores)
print("Normalized Scores:", normalized_scores)
```

##Handling Missing Values

```python
import numpy as np

#Checking missing values
arr = np.array([1, 2, np.nan, 4, 5])
print(np.isnan(arr))

#Counting missing values
print(np.sum(np.isnan(arr)))

#Replacing missing values with 0
arr[np.isnan(arr)] = 0
print(arr)
```

```python
#Replacing missing values with mean
arr = np.array([1, 2, np.nan, 4, 5])
mean_val = np.nanmean(arr)
arr[np.isnan(arr)] = mean_val
print(arr)  # [1. 2. 3. 4. 5.]
```
