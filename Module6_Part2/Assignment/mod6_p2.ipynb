**1. Introduction to NumPy and ndarray**

```python
#a) Create a 2-dimensional NumPy array of shape (3, 3) with elements of type float64. 

import numpy as np

arr_2d = np.array([[1.,2.,3.],[3.,4.,5.], [5.,6.,7.]], dtype=float)
print(arr_2d)
print(arr_2d.dtype)
```

```python
#b) Perform element-wise addition of the array with a scalar value of 10. 

result_arr_2d = arr_2d + 10
print(result_arr_2d)
print(result_arr_2d.dtype)
```

```python
#c) Perform element-wise multiplication with another 3x3 NumPy array. 

arr_2d_2 = np.array([[10.,20.,30.],[40.,50.,60.],[70.,80.,90.]],dtype=float)
result_arr_2d_2 = arr_2d * arr_2d_2
print(arr_2d)
print(arr_2d_2)
print(result_arr_2d_2)
```

```python
# d) Print the dtype, shape, and size for each of the two result arrays.
print(result_arr_2d.dtype)
print(result_arr_2d.shape)
print(result_arr_2d.size)
print(result_arr_2d_2.dtype)
print(result_arr_2d_2.shape)
print(result_arr_2d_2.size)
```

```python
#a) Create a DataFrame with missing values (NaN) in several columns.
import pandas as pd

data = {"Name":["Bob", "Alice", "Mary", "Peter"], "FavHobby":["Cooking", "Fishing", np.nan, "Surfing"], "Age":[25, np.nan,10, 20], "City":[np.nan, "Malaysia", "India", "Australia"]}
data_df = pd.DataFrame(data)
print(data_df)
```

```python
#b) Use Pandas functions to identify missing values using isna() or isnull() 

pd.isna(data_df)
# data_df.isna().sum() #count no. of missing values in each column
```

```python
#c) Use Pandas functions to drop rows with any missing data using dropna(). 

data_df.dropna(axis=0) #axis=0 row-by-row, axis=1 columns
```

```python
#d) Use Pandas functions to fill missing values with a specified value (e.g., mean or 0) using fillna(). 

data = {"Name":["Bob", "Alice", "Mary", "Peter"], "FavHobby":["Cooking", "Fishing", np.nan, "Surfing"], "Age":[25, np.nan,10, 20], "City":[np.nan, "Malaysia", "India", "Australia"]}
data_df = pd.DataFrame(data)
print(data_df)

data_df_fill = data_df.fillna("0")
print("\n",data_df_fill)
```

**3. Matplotlib Visualisation**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) Create a simple line plot showing the change in the value of a variable over time (you can generate data). 

month = pd.date_range(start='1/1/2025', periods=12, freq='M')
purchase_amt = np.abs(np.random.randn(12)*100)
purchase_trend_df = pd.DataFrame({"Month":month, "Amt":purchase_amt}) #dictionary of column and values
# Convert the float column to a string column with formatting applied
purchase_trend_df['Amt'] = purchase_trend_df['Amt'].apply(lambda x: f"${x:,.2f}")
print("Generated Monthly Purchase Data:","\n", purchase_trend_df)

plt.plot(purchase_amt)
plt.show()
```

```python
# b) Add labels to the x-axis, y-axis, and a title. 

# --- Data Generation ---
month = pd.date_range(start='1/1/2025', periods=12, freq='M')
purchase_amt = np.abs(np.random.randn(12)*100)
purchase_trend_df = pd.DataFrame({"Month":month, "Amt":purchase_amt}) #dictionary of column and values

# --- Simplest Plotting ---
purchase_trend_df.plot(x='Month', y='Amt', kind='line', marker='o')

# Adding basic labels using the pyplot interface
plt.title("Monthly Purchase Amounts in 2025")
plt.xlabel("Month")
plt.ylabel("Purchase Amount ($)")

# plt.plot()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#c) Customise the line plot by changing the colour, marker style, and line width. 

# --- Data Generation ---
month = pd.date_range (start="1/1/2025", periods=12, freq="M")
purchase_amt = np.abs(np.random.randn(12)*100)
purchase_trend_df = pd.DataFrame({"Month":month, "Amt":purchase_amt})
print(purchase_trend_df)

# --- Simplest Plotting ---
purchase_trend_df.plot(x='Month', y='Amt', kind='line', marker='s', color='red', linewidth='2')

# Adding basic labels using the pyplot interface
plt.title("Monthly Purchase Amounts in 2025")
plt.xlabel("Month")
plt.ylabel("Amt")

# plt.plot()
plt.tight_layout()

# Display the plot
plt.show()
```
