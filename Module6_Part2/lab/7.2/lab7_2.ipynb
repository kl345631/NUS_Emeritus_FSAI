```python
import pandas as pd
import os

# Define the relative file path to the folder and the file name
# Note the forward slash before the folder name
folder_path = './data_lab7.2/'
file_name = 'occupation.csv'

# Combine them to create the full path string
file_path = os.path.join(folder_path, file_name)

# read csv
occupations_df = pd.read_csv(file_path, index_col=0)

# Display the first few rows to verify it loaded correctly
print(f"DataFrame successfully loaded from: {file_path}")
print(occupations_df.head(5))
```
